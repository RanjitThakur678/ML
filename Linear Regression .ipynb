```python
# feature or data which we we will use 
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
# target which we will get from data
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
# Here we are splitting datas into test and train set and we denoted by tuple unpacking using "," 
# test size is percentage value of data we used for test , random state is number to signify the random state

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
  
```

```python
# importing Liner regression model from sklearn
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression() 
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
X_train.columns
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
```

```python
cdf
```

```python
# Predictions of our target on the basis of features or data (X_Test)

predictions = lm.predict(X_test)
predictions
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.histplot((y_test-predictions))

# if we have normally distributed residuals that mean we had correct model choice if not then we need go back and check
# the best choice of model
```

```python
# Evaluation of Regression metrics
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
import numpy as np
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```
